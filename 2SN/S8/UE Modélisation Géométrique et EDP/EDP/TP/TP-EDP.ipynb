### 1.3.2 Établissement des formules donnant la matrice de raideur associée à un élément de type quadrangle
Pour dériver la matrice de rigidité $M$ de l’élément quadrilatéral, nous commençons par définir le gradient des fonctions de forme aux coordonnées naturelles $( \xi,  \zeta)$ en $ \nabla \phi ( \xi,  \zeta)$. L’expression pour $\nabla \phi( \xi,  \zeta)$ est donnée par :

$$ \nabla \phi( \xi,  \zeta) = \left(
    \begin{array}{ll}
        \zeta - 1 & \xi - 1\\
        1 - \zeta & - \xi \\
        \zeta & \xi \\
        - \zeta & 1 - \xi \\
    \end{array}
\right)$$

Ensuite, en utilisant la formule de changement variable, nous exprimons la matrice de rigidité $M$ comme :

$$M_{ij} =  \displaystyle \int_0^1 \int_0^1 {\nabla \phi_i^T (J_\phi^T J_\phi)^{-1} \nabla \phi_j |J_\phi|}\, \mathrm{d}\xi \, \mathrm{d}\zeta$$

avec $ \nabla \phi_i$ la i-ème ligne de $ \nabla \phi$, $J_\phi$ la matrice jacobienne de la transformation de l’élément de référence à l’élément physique et $|J_\phi|$ le déterminant de $J_\phi$.

Nous évaluons ensuite les intégrales pour obtenir $M$. Pour ce faire, nous définissons d’abord

$$(J_\phi^T J_\phi)^{-1} =
\begin{pmatrix}
a & b \\
b & c \\
\end{pmatrix}$$

comme l’inverse de la matrice jacobienne symétrique de $\phi$. On peut alors calculer les coefficients de la matrice de rigidité $M$.

Par exemple, le calcul du coefficient $M_{33}$ s'effectue de la manière suivante :
$$ \displaystyle \int_0^1 \int_0^1 {\left(\begin{array}{ll} \xi & \zeta\end{array}\right)\begin{pmatrix}a & b \\b & c \\\end{pmatrix}\left(\begin{array}{ll} \xi\\ \zeta\end{array}\right)|J_\phi|}\mathrm{d}\xi \, \mathrm{d}\zeta = |J_\phi| \int_0^1 \int_0^1 {\left(\begin{array}{ll} \xi a + \zeta b & \xi b + \zeta c\end{array}\right)\left(\begin{array}{ll} \xi\\ \zeta\end{array}\right)}\mathrm{d}\xi \, \mathrm{d}\zeta\\ = |J_\phi| \int_0^1 \int_0^1 {(\xi^2 a + 2\xi\zeta b + \zeta^2 c)}\mathrm{d}\xi \, \mathrm{d}\zeta = \int_0^1 {[\frac{1}{3}\xi^3 a + \xi^2\zeta b + \xi\zeta^2 c]^1_0}\mathrm{d}\zeta \\ = |J_\phi| \int_0^1 {(\frac{1}{3} a + \zeta b + \zeta^2 c})\mathrm{d}\zeta = |J_\phi| [\frac{1}{3} a \zeta + \frac{1}{2}\zeta^2 b + \frac{1}{3}\zeta^3 c]^1_0 \\ = |J_\phi| (\frac{1}{3} a + \frac{1}{2} b + \frac{1}{3} c) = \frac{|J_\phi|}{6}(2a + 3b + 2c)$$

En calculant ainsi tous les coefficients de $M$ on obtient : 

$$ M = \frac{|J_\phi|}{6}
\left(
    \begin{array}{ll}
        2a + 3b + 2c & -2a + c & -a -3b - c & a - 2c\\
        -2a + c & 2a - 3b + 2c & a - 2c & - a + 3b - c \\
        -a -3b - c & a - 2c & 2a + 3b + 2c & -2a + c \\
        a - 2c & - a + 3b - c & -2a + c & 2a - 3b + 2c
    \end{array}
\right)$$

Il est important de noter que la matrice de rigidité $M$ est une matrice définie symétrique et positive, ce qui caractérise la rigidité de l’élément..

### Mise sous forme variationelle et vérification des hypothèses de Lax-Milgram
Soit le problème 

$$
\begin{cases}
    -\Delta u(x,y) + c_0 u(x,y) = f(x, y) \quad \text{ sur } \Omega, \\
    u(x,y) = 0 \quad \quad \qquad \qquad \text{ sur } \partial \Omega, \\
\end{cases}
$$

avec $c_0 > 0$ une constante et $f \in L^2(\mathbb{\Omega})$.

**1. Réécriture dans le langage des distribuitions**

$$
u \in H^2(\Omega) \cap H_0^1(\Omega), 
\begin{cases}
    -\Delta u + c_0 u = f \quad \text{ sur } \Omega, \\
    \gamma_0 (u) = 0 \quad \qquad \text{ sur } \partial \Omega, \\
\end{cases}
$$

**2. Application du théoreme de Green**

$$
u \in H^2(\Omega) \cap H_0^1(\Omega), \forall v \in L^2(\Omega), \int_{\Omega} -\Delta u v dx + c_0 \int_{\Omega} u v dx = \int_{\Omega} fv dx
$$

$$
u \in H_0^1(\Omega), \forall v \in H_0^1(\Omega), \int_{\Omega} \nabla u \nabla v dx + c_0 \int_{\Omega} u v dx - \int_{\partial \Omega} \gamma_1 (u) \gamma_0 (v) d\sigma = \int_{\Omega} fv dx
$$

$$
u \in H_0^1(\Omega), \forall v \in H_0^1(\Omega), \int_{\Omega} \nabla u \nabla v dx + c_0 \int_{\Omega} u v dx = \int_{\Omega} fv dx
$$

car $\gamma_0 (v) = 0$ sur $\partial \Omega$

**3. Détermination de $a$ et $l$**

Soient $u \in H_0^1(\Omega)$ et $v \in H_0^1(\Omega)$ on a

$$
a(u,v) = \int_{\Omega} \nabla u \nabla v dx + c_0 \int_{\Omega} u v 
$$

bien bilinéaire (par linéarité de l'intégrale et bilinéarité des produits scalaire et usuel)

$$
l(v) = \int_{\Omega} f v dx
$$

bien linéaire (par linéarité de l'intégrale)

**4. Vérification des hypothèses de Lax-Milgram (pour la norme $| \cdot |_{0,1}$)**

Soient $u \in H_0^1(\Omega)$ et $v \in H_0^1(\Omega)$ montrons

 - Continuité de $a$ :
 
 $$
 |a(u,v)| \newcommand*{\defeq}{\stackrel{\text{def}}{=}} |<u,v>_{0,1} + c_0<u,v>_{L^2(\Omega)}| \\
 \newcommand*{\defeq}{\stackrel{\text{$c_0>0$,ineg. triang.,C-S}}{\leq}} |u|_{0,1} |v|_{0,1} + c_o ||u||_{L^2(\Omega)} ||v||_{L^2(\Omega)} \\
 \newcommand*{\defeq}{\stackrel{\text{def}}{\leq}} ||u||_{H^1(\Omega)} ||v||_{H^1(\Omega)} + c_0 ||u||_{H^1(\Omega)} ||v||_{H^1(\Omega)} \\
 = (1 + c_0) ||u||_{H^1(\Omega)} ||v||_{H^1(\Omega)}
 $$
 
 Donc, par définition, $a$ est continue sur $H_0^1(\Omega) \times H_0^1(\Omega)$ pour la norme $|| \cdot ||_{H^1(\Omega)}$.
 
 Par équivalence des normes $|| \cdot ||_{H^1(\Omega)}$ et $| \cdot |_{0,1}$ sur $H_0^1(\Omega)$, $a$ est ainsi continue sur $H_0^1(\Omega) \times H_0^1(\Omega)$ pour la norme $| \cdot |_{0,1}$
 
- Coercivité de $a$ :
  
  $$
  |a(u,v)| \newcommand*{\defeq}{\stackrel{\text{def}}{=}} \left| |u|_{0,1}^2 + c_0 ||u||^2_{L^2(\Omega)} \right| \newcommand*{\defeq}{\stackrel{c_0 > 0}{=}} |u|_{0,1}^2 + c_0 ||u||^2_{L^2(\Omega)} \geq 1 |u|_{0,1}^2
  $$
  
  Donc, par définition, $a$ est coercive sur $H_0^1(\Omega) \times H_0^1(\Omega)$ pour la norme $| \cdot |_{0,1}$
  
- Continuité de $l$ :
   
   $$
   |l(v)| \newcommand*{\defeq}{\stackrel{\text{def}}{=}} |<f,v>_{L^2(\Omega)}| \\
   \newcommand*{\defeq}{\stackrel{C-S}{\leq}} ||f||_{L^2(\Omega)} ||v||_{L^2(\Omega)} \\
   \newcommand*{\defeq}{\stackrel{\text{Poincaré, $\exists C_p(\Omega)>0$}}{\leq}} ||f||_{L^2(\Omega)} C_p(\Omega)|v|_{0,1}
   $$
   
   Donc, par définition, $l$ est continue sur $H_0^1(\Omega)$ pour la norme $| \cdot |_{0,1}$.
   
   $H_0^1(\Omega)$ étant de plus un espace de Hilbert pour le produit scalaire réduit $<,>_{0,1}$, d'après le théorème de Lax-Milgram, le problème admet une unique solution sur $H_0^1(\Omega)$.

### Forme discrète de la formulation variationnelle
On note $V = H^1_0(\Omega)$\
On introduit $V_n=(Vect(\eta_i))_{1\leq i\leq n}$ avec $n=dim(V_n)$\
Donc $\forall v_n \in V_n, \exists!(\alpha_i)_{1\leq i\leq n} \in \mathbb{R^n}|v_n = \sum_{i=1}^{n}{\alpha_i \eta_i}$\
Donc on a : $u_n = \sum_{i=1}^{n}{\alpha_i\eta_i}$, avec $(\alpha_i)_{1\leq i\leq n} \in \mathbb{R^n}$, solution de $(P_{FV_n}) \Leftrightarrow \forall v_n \in V_n, a(u_n,v_n) = l(v_n) \Leftrightarrow \forall i \in  [\![1;n]\!] , a(u_n, \eta_i) = l(\eta_i) \Leftrightarrow \forall i \in [\![1;n]\!], \sum_{j=1}^{n}{\alpha_j a(\eta_j, \eta_i)} = l(\eta_i) \Leftrightarrow Ax=b$
Avec :\
$A \in \mathcal{M}_n(\mathbb{R})$ telle que $\forall (i,j) \in [\![1;n]\!]^2, A_{ij} = a(\eta_j,\eta_i) = \int_{\Omega}{\nabla\eta_i^\top \nabla \eta_j}\mathrm{d}x + c_0\int_{\Omega}{\eta_i \eta_j}\mathrm{d}x$\
$b \in \mathbb{R^n}$ tel que $\forall i \in [\![1;n]\!], b_i = l(\eta_i) = \int_{\Omega}{f\eta_i}\mathrm{d}x$\
$x = (\alpha_i)_{1\leq i \leq n}$\
Montrons à présent que A définie positive :\
Soit $x \in \mathbb{R^n}\backslash\lbrace{0}\rbrace$.$x^\top A x = \sum_{i=1}^{n}\sum_{j=1}^{n}{x_i x_j A_{ij}} = \sum_{i=1}^{n} \sum_{j=1}^{n}{x_i x_j a(\eta_j, \eta_i)} = a(\sum_{j=1}^{n}{x_j \eta_j},\sum_{i=1}^{n}{x_i \eta_i}) = a(y,y) \geq \alpha \| y \|^2_V$ (avec $y = \sum_{i=1}^{n}{x_i \eta_i}$ et $\alpha$ une constante de coercivité de $a$). Donc A est semi-définie positive.
De plus, on a : $x^\top A x = 0 \Rightarrow \| y \|_V = 0 \Rightarrow y = 0 \Rightarrow \forall i \in [\![1;n]\!], x_i = 0 \Rightarrow x = 0$\
Donc A est définie positive et, a fortiori, est inversible.\
Ainsi, le système d'équations $Ax = b$, tel que défini ci-dessus, admet une unique solution.

### Stratégie de l'élément de référence
La prise en compte du terme $c_0\int_\Omega {\eta_i \eta_j} \mathrm{d}x$ se fait par ajout de sa matrice de raideur associée à la matrice de raideur du terme $\int_\Omega {\nabla \eta_i^\top \nabla \eta_j} \mathrm{d}x$, et ce pour chaque triangle du maillage (sachant qu'il n'y a pas de quadrangles dans cette partie).\
Pour un triangle de référence aux sommets $(0,0)$,$(1,0)$ et $(0,1)$, les fonctions de base correspondant à ces sommets sont définies pour tout couple $(\xi,\zeta)$ par $\phi = \left(\begin{array}{ll} 1-\xi -\zeta & \xi & \zeta \end{array}\right)$.\
Ensuite, les formules fournies dans l'annexe permettent de relier $\phi$ et $\eta$ de la manière suivante : $\int_{T}{\eta_i \eta_j} \mathrm{d}x$ = $|J_{\phi_T}|\int_{T}{\phi_i \phi_j} \mathrm{d}x$\
En calculant ainsi tous les coefficients de $M$ on obtient : 

$$ M = \frac{|J_\phi|}{24}
\left(
    \begin{array}{ll}
        2 & 1 & 1\\
        1 & 2 & 1\\
        1 & 1 & 2
    \end{array}
\right)$$

```python
def maillage(n):
#
# Une discretisation possible d'une EDP elliptique sur le domaine ]0,1[ x ]0,1[
# Le carre [0,1]x[0,1] est maille uniquement avec des triangles; 
# Les conditions limites sont de type Dirichlet uniquement   => neumann  =[];
#
# Entrees :
# n : nombre de points par cote du care => Npts points de discretisation au
# total
#
# Sorties :
# coordinates : matrice a deux colonnes. Chaque ligne contient les 
# coordonnes 2D d'un des points de la discretisation. Ces sommets seront 
# identifies a l'indice de la ligne correspondante dans la matrice
# coordinates.
# elements3 : matrice a trois colonnes. Chaque ligne contient les indices 
# des sommets d'un element triangle, dans le sens antihoraire. 
# dirichlet : vecteur colonne des indices des sommets de la frontiere de
# Dirichlet.
# neumann : matrice a deux colonnes. Chaque ligne contient les indices 
# des deux sommets d'une arete de la frontiere de Neumann.
# (neumann est vide sur cet exemple)
#
##################################################################################
    h=1/(n-1)
    npoin       = n*n ; 
    nelem       = 2*(n-1)*(n-1) ;
    coordinates = np.zeros((npoin,2)); 
    elements3   = (np.zeros((nelem,3))).astype(int) ;
    neumann     = [];
    dirichlet=(np.zeros((4*n-4,1))).astype(int)
    # Coordonnees et connectivites :
    e = -1 ; 
    p = -1 ;
    x=np.zeros((n+1,1))
    x[n,0]=1.
    for l in range (n+1):
        x[l,0]=l*h
    for j in range (n):
            for i in range(n):
                p = p + 1  
                coordinates[p,0] = x[i,0]  
                coordinates[p,1] = x[j,0] 
                if ((i != n-1) & (j != n-1)):
                    p1 = p
                    p2 = p1 + 1 
                    p3 = p1 + n 
                    p4 = p2 + n 
                    e = e + 1 
                    elements3[e,0] = p1 
                    elements3[e,1] = p2 
                    elements3[e,2] = p3 
                    e = e + 1
                    elements3[e,0] = p4 
                    elements3[e,1] = p3 
                    elements3[e,2] = p2 
    #Liste des sommets de la frontiere de Dirichlet:
    p=-1
    for j in range(n):
        p=p+1
        dirichlet[p,0] = j  
    for j in range(n*2-1,n*(n-1),n):
        p=p+1
        dirichlet[p,0] = j 
    for j in range(n*n-1,n*n-n-1,-1):
        p=p+1
        dirichlet[p,0] = j 
    for j in range(n*n-2*n,n-1,-n):
        p=p+1
        dirichlet[p,0] = j 

    return coordinates, elements3, dirichlet, neumann
```

```python
def show(coordinates,u):
#
# Fonction d'affichage de la solution u sur le maillage defini par
# elements3, coordinates.
#
# Entrees:
# elements3 : matrice a trois colonnes contenant les elements triangles
# de la discretisation, identifies par les indices de leurs trois
# sommets.
# coordinates : matrice a deux colonnes contenant les coordonnes 2D des
# points de la discretisation.
# u : vecteur colonne de longueur egale au nombre de lignes de
# coordinates contenant les valeurs de la solution a afficher aux
# points de la discretisation.
#
# Sorties : Aucune, mais la fonction doit s'afficher dans une figure.
##########################################################################
    ax= plt.figure().add_subplot(projection='3d')
    ax.plot_trisurf(coordinates[:,0],coordinates[:,1],u,linewidth=0.2,antialiased=True)
    plt.show()
```

```python
def calculate_rigidity_matrix_triangle(vertices):
    r"""Calculates the rigidity matrix
    Input: vertices of the triangle
    Output: M matrix, alpha float
    """
    # Get the vertices coordinates and define alpha
    x = np.array(vertices)[:, 0]
    y = np.array(vertices)[:, 1]
    alpha = (x[1]-x[0])*(y[2]-y[0]) - (x[2]-x[0])*(y[1]-y[0])
    
    M = np.zeros((3,3))
    # Calculate the coefficients of the matrix M
    for i in range(3):
        for j in range(3):
            # Base functions over the triangle as defined in the subject
            grad_eta_i = (1/alpha) * np.array([y[(i+1)%3]-y[(i+2)%3], x[(i+2)%3]-x[(i+1)%3]])
            grad_eta_j = (1/alpha) * np.array([y[(j+1)%3]-y[(j+2)%3], x[(j+2)%3]-x[(j+1)%3]])
            
            # Get the coefficient (i,j) of M
            # 2|T| = alpha and M = |T|*grad_eta_i^T * grad_eta_j
            M[i,j] = (alpha/2) * (np.transpose(grad_eta_i).dot(grad_eta_j))
    
    return M, alpha
```

```python
def assembling_triangle(coordinates, elements3, dirichlet):
    r"""Assemblage de la matrice A dans le cas d’un maillage 
    constitu´e uniquement d’´el´ements triangles
    Input: coordinates matrix, elements3 matrix, dirichlet matrix
    Output: A matrix, b vector, U_h matrix
    """
    n_triangles = len(coordinates[elements3])
    triangles = coordinates[elements3]
    n_vertices = len(coordinates)
    A = np.zeros((n_vertices, n_vertices))
    b = np.zeros(n_vertices)
    
    # Defines the funtion for the second member
    def f(x,y):
        return 1
    
    # Defines the function for the Dirichlet conditions
    def u_d(x,y):
        return 1
    
    # Calculates the matrix A and the vector b with a triangle mesh
    for k in range(n_triangles):
        vertices_tr = triangles[k]
        # Get the rigity matrix with respect with the given triangle
        M_k, alpha = calculate_rigidity_matrix_triangle(vertices_tr)
        # Calculates the barycenter of the triangle (center of gravity)
        xg, yg = np.sum(vertices_tr/3, axis=0)
        # Approximation of the quadrature based on the center of grabity of the triangles
        b_k = alpha/6 * f(xg, yg)
        
        # Defines a dictionary to match each element of A with the calculated M_k
        i,j,l = elements3[k,:]
        dico = {0:i, 1:j, 2:l}
        for key1 in dico:
            b[dico[key1]] += b_k
            for key2 in dico:
                A[dico[key1],dico[key2]] += M_k[key1, key2]
    
    # For each element in the Dirichlet frontier calculates the value of U_h
    U_h = np.zeros(len(b))
    for v in dirichlet:
        vertex = coordinates[v][0]
        U_h[v] = u_d(vertex[0], vertex[1])
    
    # Update b
    b = b - A.dot(U_h)
    
    # For each coordinate that is not in the Dirichlet (get the inner part of the domain)
    vertices_not_dirichlet = np.setdiff1d([i for i in range(len(coordinates))], dirichlet[:,0])
    
    # Updates the values of A, b and U_h
    A_aux = A[vertices_not_dirichlet]
    A = A_aux[:,vertices_not_dirichlet]
    b = b[vertices_not_dirichlet]
    U_h_aux = (np.linalg.inv(A)).dot(b)
    U_h[vertices_not_dirichlet] = U_h_aux 
    
    return A, b, U_h
```

```python
def calculate_rigidity_matrix_square(vertices):
    r"""Calculates the rigidity matrix of a square
    Input: vertices of the square
    Output: M matrix, alpha vector
    """
    
    # Get the coordinates of each vertex
    x = np.array(vertices)[:, 0]
    y = np.array(vertices)[:, 1]
    # Calculates the matrix J and its inverse
    J = np.array([[(x[1]-x[0]), (x[3]-x[0])], [(y[1]-y[0]), (y[3]-y[0])]])
    alpha = (x[1]-x[0])*(y[2]-y[0]) - (x[2]-x[0])*(y[1]-y[0])
    J_inv = np.linalg.inv(J.dot(np.transpose(J)))
    # To calculate the values of M
    a,b,c = J_inv[0,0], J_inv[0,1], J_inv[1,1]
    
    M = np.zeros((4,4))
    
    # Line 1
    M[0,0] = (2*a + 3*b + 2*c)
    M[0,1] = (-2*a + c)
    M[0,2] = (-a -3*b - c)
    M[0,3] = (a - 2*c)
    
    # Line 2
    M[1,1] = (2*a - 3*b + 2*c)
    M[1,2] = (a - 2*c)
    M[1,3] = (-a + 3*b - c)
    M[1,0] = (-2*a + c)
    
    # Line 3
    M[2,0] = (-a -3*b - c)
    M[2,1] = (a - 2*c)
    M[2,2] = (2*a + 3*b + 2*c)
    M[2,3] = (-2*a + c)
    
    # Line 4
    M[3,0] = (a - 2*c)
    M[3,1] = (-a + 3*b - c)
    M[3,2] = (-2*a + c)
    M[3,3] = (2*a - 3*b + 2*c)
    
    # Using the formula given in the subject
    M = np.linalg.det(J)/6 * M
    
    return M, alpha
    
    
    
    
```

```python
def assembling_mixed(coordinates, elements3, elements4, dirichlet, neumann):
    r"""Assemblage de la matrice A dans le cas d’un maillage 
    constitu´e d’´el´ements triangles et quadrangles
    Input: coordinates matrix, elements3 matrix, elements4 matrix, dirichlet matrix, neumann matrix
    Output: A matrix, b vector, U_h matrix
    """
    # Preprocess Dirichlet
    dirichlet = dirichlet.reshape(-1, 1)
    n_triangles = len(coordinates[elements3])
    triangles = coordinates[elements3]
    n_squares = len(coordinates[elements4])
    squares = coordinates[elements4]
    n_vertices = len(coordinates)
    A = np.zeros((n_vertices, n_vertices))
    b = np.zeros(n_vertices)
    
    # Function f second member
    def f(x,y):
        return 1
    
    # Dirichlet condition
    def u_d(x,y):
        return 1
    
    # Neumann condition
    def g(x,y):
        return 1
    
    # Calculates the triangles
    for k in range(n_triangles):
        vertices_tr = triangles[k]
        # Calculates the rigity matrix for triangles
        M_k, alpha = calculate_rigidity_matrix_triangle(vertices_tr)
        xg, yg = np.sum(vertices_tr/3, axis=0)
        b_k = alpha/6 * f(xg, yg)
        
        # Updates the values of A and b
        i,j,l = elements3[k,:]
        dico = {0:i, 1:j, 2:l}
        for key1 in dico:
            b[dico[key1]] += b_k
            for key2 in dico:
                A[dico[key1],dico[key2]] += M_k[key1, key2]
                
    # Calculates the squares            
    for k in range(n_squares):
        vertices_sq = squares[k]
        # Calculates the rigity matrix for quadrangles
        M_k, alpha = calculate_rigidity_matrix_square(vertices_sq)
        # Calculates the center o gravity of the quadrangles
        xg, yg = np.sum(vertices_tr/4, axis=0)
        b_k = alpha/4 * f(xg, yg)
        
        # Updates the matrix A and b
        i,j,l,m = elements4[k,:]
        dico = {0:i, 1:j, 2:l, 3:m}
        for key1 in dico:
            b[dico[key1]] += b_k
            for key2 in dico:
                A[dico[key1],dico[key2]] += M_k[key1, key2]
    
    # Applies the Dirichlet condition
    U_h = np.zeros(len(b))
    for v in dirichlet:
        vertex = coordinates[v][0]
        U_h[v] = u_d(vertex[0], vertex[1])
        
    # Updates b according to the Dirichlet condition
    b = b - A.dot(U_h)
    
    # Applies the Neumann condition
    for i,j in neumann:
        v1, v2 = coordinates[i], coordinates[j]
        # Calculates the center of the segment
        x_c, y_c = np.sum(v1)/2, np.sum(v2)/2
        d = np.linalg.norm(v1 - v2)
        # Calculates the limit and add it to b for the Neumann condition
        boundary_measure = g(x_c, y_c)*(d/2)
        b[i], b[j] = b[i] + boundary_measure, b[j] + boundary_measure
    
    # Updates the values of A, b and U_h based on the inner points
    vertices_not_dirichlet = np.setdiff1d([i for i in range(len(coordinates))], dirichlet[:,0])
    
    A_aux = A[vertices_not_dirichlet]
    A = A_aux[:,vertices_not_dirichlet]
    b = b[vertices_not_dirichlet]
    
    U_h_aux = (np.linalg.inv(A)).dot(b)
    U_h[vertices_not_dirichlet] = U_h_aux 
    
    return A, b, U_h
```

```python
def Mc():
    r"""Defines the matrix M
    Output: Mat matrix
    """
    Mat = np.zeros((3,3))

    # Line 1
    Mat[0,0] = 2
    Mat[0,1] = 1
    Mat[0,2] = 1

    # Line 2
    Mat[1,0] = 1
    Mat[1,1] = 2
    Mat[1,2] = 1

    # Line 3
    Mat[2,0] = 1
    Mat[2,1] = 1
    Mat[2,2] = 2
    
    return Mat
```

```python
def assembling_triangle_complements(coordinates, elements3, dirichlet):
    r"""Assemblage de la matrice A dans le cas d’un maillage 
    constitu´e uniquement d’´el´ements triangles
    Input: coordinates matrix, elements3 matrix, dirichlet matrix
    Output: A matrix, b vector, U_h matrix
    """
    n_triangles = len(coordinates[elements3])
    triangles = coordinates[elements3]
    n_vertices = len(coordinates)
    A = np.zeros((n_vertices, n_vertices))
    b = np.zeros(n_vertices)
    
    # Defines c0 > 0
    c0 = 1
    
    # Defines f second member
    def f(x,y):
        return 1
    
    # Defines Dirichlet conditions
    def u_d(x,y):
        return 0
    
    # Calculates the triangles 
    for k in range(n_triangles):
        vertices_tr = triangles[k]
        # Calculates the rigity matrix for the triangle
        M_k, alpha = calculate_rigidity_matrix_triangle(vertices_tr)
        # Calculates the matrix Mt_k based on the predefined matrix Mc()
        Mt_k = c0*(alpha/24)*Mc()
        xg, yg = np.sum(vertices_tr/3, axis=0)
        # Calculates the approximation based on the center of gravity of the triangle
        b_k = alpha/6 * f(xg, yg)
        
        # Updates the values of A and b
        i,j,l = elements3[k,:]
        dico = {0:i, 1:j, 2:l}
        for key1 in dico:
            b[dico[key1]] += b_k
            for key2 in dico:
                A[dico[key1],dico[key2]] += M_k[key1, key2] + Mt_k[key1, key2]
    
    # Applies the Dirichlet conditions
    U_h = np.zeros(len(b))
    for v in dirichlet:
        vertex = coordinates[v][0]
        U_h[v] = u_d(vertex[0], vertex[1])
    
    # Updates b based on the Dirichlet conditions
    b = b - A.dot(U_h)
    
    # Updates the inner points of the domain
    vertices_not_dirichlet = np.setdiff1d([i for i in range(len(coordinates))], dirichlet[:,0])
    
    A_aux = A[vertices_not_dirichlet]
    A = A_aux[:,vertices_not_dirichlet]
    b = b[vertices_not_dirichlet]
    
    U_h_aux = (np.linalg.inv(A)).dot(b)
    U_h[vertices_not_dirichlet] = U_h_aux 
    
    return A, b, U_h
```
